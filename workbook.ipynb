```python
import numpy as np 
import math 
import random 
import matplotlib.pyplot as plt 
```

```python
# exponent via math 
print(math.exp(-3))
# exponent via numpy 
print(np.exp(-3))
```

```python
random.seed(30)
x = np.arange(-10,11)
def sigmoid(x):
    return 1 / ( 1 + np.exp(-x))
def softmax(lst):
    lstNew = []
    eSum = np.sum(np.exp(np.array(lst)))
    for i in lst:
        xNew = np.exp(i) / eSum
        lstNew.append(xNew)
    return lstNew
```

```python
ySigmoid = [sigmoid(n) for i,n in enumerate(x)]
ySoftmax = softmax(x)
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.plot(x,ySigmoid,c='red')
ax.plot(x,ySoftmax,c='blue')
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
# ax.spines['bottom'].set_position('center')
ax.spines['top'].set_color('none')
ax.spines['left'].set_smart_bounds(True)
# ax.spines['bottom'].set_smart_bounds(True)
```
